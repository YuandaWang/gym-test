```python
import gym
import matplotlib.pyplot as plt
env = gym.make('Pendulum-v0')
env.reset()
env.render()
s1 = []
s2 = []
s3 = []
for i in range(3):
    action = env.action_space.sample()
    action = [0]
    s, r, d, info = env.step(action)
    s1.append(s[0])
    s2.append(s[1])
    s3.append(s[2])
    print s, r, action
    
    env.render()
```

```python
plt.plot(s1, color='red')
plt.plot(s2, color='blue')
plt.plot(s3, color='green')
plt.show()
```
